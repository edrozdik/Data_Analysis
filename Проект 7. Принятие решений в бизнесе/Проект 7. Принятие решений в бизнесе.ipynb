### –ü—Ä–∏–º–µ–Ω–∏–º —Ñ—Ä–µ–π–º–≤–æ—Ä–∫ ICE –¥–ª—è –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏–∏ –≥–∏–ø–æ—Ç–µ–∑.  –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –∏—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞.  

$$
ICE = \frac{impact \times confidence}{efforts}
$$

````python
hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
````

````python
# hypothesis[['hypothesis', 'ICE']]: 
# –ò–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ hypothesis –≤—ã–±–∏—Ä–∞—é—Ç—Å—è –¥–≤–∞ —Å—Ç–æ–ª–±—Ü–∞ ‚Äî hypothesis –∏ ICE. 
# –≠—Ç–æ —Å–æ–∑–¥–∞—ë—Ç –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –∫–æ—Ç–æ—Ä—ã–π —Å–æ—Å—Ç–æ–∏—Ç —Ç–æ–ª—å–∫–æ –∏–∑ —ç—Ç–∏—Ö –¥–≤—É—Ö —Å—Ç–æ–ª–±—Ü–æ–≤.
# .sort_values(by='ICE', ascending=False): –ù–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç—Å—è –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º –≤ —Å—Ç–æ–ª–±—Ü–µ ICE. 

hypothesis_ICE = hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).round(2)
````

–†–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç–æ–¥–∞ ICE –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç:

–ü—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω–æ—Å—Ç—å –∑–∞–¥–∞—á–∏ ‚Äî –∑–∞–¥–∞—á–∏ —Å –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–º ICE-–∑–Ω–∞—á–µ–Ω–∏–µ–º —Å—á–∏—Ç–∞—é—Ç—Å—è –±–æ–ª–µ–µ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã–º–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è, —Ç–∞–∫ –∫–∞–∫ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—é—Ç –≤—ã—Å–æ–∫–∏–π —ç—Ñ—Ñ–µ–∫—Ç –ø—Ä–∏ –º–µ–Ω—å—à–∏—Ö —É—Å–∏–ª–∏—è—Ö.

### –ü—Ä–∏–º–µ–Ω–∏–º —Ñ—Ä–µ–π–º–≤–æ—Ä–∫ RICE –¥–ª—è –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏–∏ –≥–∏–ø–æ—Ç–µ–∑. –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –∏—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞.
$$
ICE = \frac{reach \times impact \times confidence}{efforts}
$$

–§—Ä–µ–π–º–≤–æ—Ä–∫ **RICE** —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç:  
**Reach** ‚Äî —Å–∫–æ–ª—å–∫–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∑–∞—Ç—Ä–æ–Ω–µ—Ç –∏–∑–º–µ–Ω–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã —Ö–æ—Ç–∏—Ç–µ –≤–Ω–µ—Å—Ç–∏;  
**Impact** ‚Äî –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∏—Ö –æ–ø—ã—Ç –∏ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–∏–µ –æ—Ç –ø—Ä–æ–¥—É–∫—Ç–∞;  
**Confidence** ‚Äî –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤—ã —É–≤–µ—Ä–µ–Ω—ã, —á—Ç–æ —ç—Ç–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–º–µ–Ω–Ω–æ —Ç–∞–∫;   
**Efforts** ‚Äî —Å–∫–æ–ª—å–∫–æ —Å—Ç–æ–∏—Ç –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å —ç—Ç—É –≥–∏–ø–æ—Ç–µ–∑—É. 

RICE - —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è ICE, –∫–æ—Ç–æ—Ä–∞—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏–∏ –∑–∞–¥–∞—á. –û–Ω –ø–æ–º–æ–≥–∞–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å, –∫–∞–∫–∏–µ –∑–∞–¥–∞—á–∏ –ø—Ä–∏–Ω–µ—Å—É—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–æ–ª—å–∑—É –ø—Ä–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö –∑–∞—Ç—Ä–∞—Ç–∞—Ö –≤—Ä–µ–º–µ–Ω–∏ –∏ —Ä–µ—Å—É—Ä—Å–æ–≤.

````python
hypothesis['RICE'] = hypothesis['impact'] * hypothesis['reach'] * hypothesis['confidence'] / hypothesis['efforts']
````

````python
hypothesis_RICE = hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False).round(2)
````

````python
# —Å–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø —Ç–µ—Å—Ç–∞
datesGroups = orders_cleaned[['date','group']].drop_duplicates() 
````

````python
# –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø–æ–ª—É—á–∏–º –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–µ –ø–æ –¥–Ω—è–º –¥–∞–Ω–Ω—ã–µ –æ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è—Ö –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞: 
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group']) 
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
````

### –ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º.

````python
# —Ç–∞–±–ª–∏—Ü—ã —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –∑–∞–∫–∞–∑–∞–º –Ω–∞ –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
orders_by_users_a = orders_cleaned[orders_cleaned['group']=='A'].groupby('visitor_id')['transaction_id'].nunique().reset_index()
orders_by_users_a.columns=['visitors_id','orders']

orders_by_users_b = orders_cleaned[orders_cleaned['group']=='B'].groupby('visitor_id')['transaction_id'].nunique().reset_index()
orders_by_users_b.columns=['visitors_id','orders'] 

display(orders_by_users_a.head()) 
display(orders_by_users_b.head())  
````

–û–±—ä—è–≤–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º —Å –∑–∞–∫–∞–∑–∞–º–∏ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤, –∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –±–µ–∑ –∑–∞–∫–∞–∑–æ–≤ ‚Äî –Ω—É–ª–∏.

````python
sampleA = pd.concat(
    [
        orders_by_users_a['orders'],
        pd.Series(0, 
                  index=np.arange
                  (visitors[visitors['group']=='A']['visitors'].sum() - len(orders_by_users_a['orders'])
                  ),
                  name='orders'
                 )
    ],
    axis=0)

sampleB = pd.concat(
    [
        orders_by_users_b['orders'],
        pd.Series
        (0,
         index=np.arange(
             visitors[visitors['group']=='B']['visitors'].sum() - len(orders_by_users_b['orders'])
         ),
         name='orders'
        )
    ],
    axis=0)
````

**H0:** –†–∞–∑–ª–∏—á–∏–π –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç

**H1:** –†–∞–∑–ª–∏—á–∏—è –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å

<div class="alert alert-success">
  
  <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</b>
    
  ‚úîÔ∏è –ù—É–ª–µ–≤–∞—è –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã. 

–°–∞–º–∏ –≥–∏–ø–æ—Ç–µ–∑—ã –æ–±—ã—á–Ω–æ –æ–±–æ–∑–Ω–∞—á–∞—é—Ç —Ç–∞–∫:
- H0 - –ù–µ—Ç  —Ä–∞–∑–ª–∏—á–∏–π ....
- H1 - –ï—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏—è ....

  –ø–æ–ø—Ä–∞–≤–∏–ª–∞ 
</div>

````python
result_1 = stats.mannwhitneyu(sampleA, sampleB)[1] 
#–∑–∞–¥–∞–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ aplha
alpha = 0.05
#–≤–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ p-value
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ p-value : {result_1}')
# —É—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä —Å –≤—ã–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏ —Å –æ—Ç–≤–µ—Ç–æ–º
if result_1 < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")  
print('–°—Ä–µ–¥–Ω—è—è –∫–æ–Ω–≤–µ—Ä—Å–∏—è –≥—Ä—É–ø–ø—ã –ê: {:.1%}'.format(sampleA.mean()))
print('–°—Ä–µ–¥–Ω—è—è –∫–æ–Ω–≤–µ—Ä—Å–∏—è –≥—Ä—É–ø–ø—ã B: {:.1%}'.format(sampleB.mean()))  
print('–û—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –í –∫ –ê: {0:.3f}'.format(sampleB.mean()/sampleA.mean()-1)) 
````

P-value –º–µ–Ω—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è alpha - –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –≤–µ—Ä–Ω–æ —É—Ç–≤–µ–∂–¥–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ –∫–æ–Ω–≤–µ—Ä—Å–∏—è—Ö –µ—Å—Ç—å.

–∫–æ–ª-–≤–æ > 2, –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è more_orders  
—Å—É–º–º–∞ > 30000, –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è to_expensive_order

````python
# —Å–æ–±–∏—Ä–∞–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ 
many_orders = pd.concat(
    [
        orders_by_users_a[orders_by_users_a['orders'] > more_orders]['visitors_id'],
        orders_by_users_b[orders_by_users_b['orders'] > more_orders]['visitors_id'],
    ],
    axis=0,
)

# —Å–æ–±–∏—Ä–∞–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ —Å–ª–∏—à–∫–æ–º –¥–æ—Ä–æ–≥–∏–º–∏ –∑–∞–∫–∞–∑–∞–º–∏
expensive_orders = orders_cleaned[orders_cleaned['revenue'] > to_expensive_order]['visitor_id']

# –æ–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å—ë –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Ñ—Ä–µ–π–º 
abnormal_users = (
    pd.concat([many_orders, expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormal_users.head(5))
print(' ')
print('–ö–æ–ª-–≤–æ –∞–Ω–∞–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: ', abnormal_users.shape[0])
````

````python
#–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ –≥—Ä—É–ø–ø–µ –ê 
sampleAFiltered = pd.concat(
    [
        orders_by_users_a[
            np.logical_not(orders_by_users_a['visitors_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors[visitors['group']=='A']['visitors'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

#–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ –≥—Ä—É–ø–ø–µ –í
sampleBFiltered = pd.concat(
    [
        orders_by_users_b[
            np.logical_not(orders_by_users_b['visitors_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors[visitors['group']=='B']['visitors'].sum() - len(orders_by_users_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)  
````

````python
print('P-value:','{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ —Ä–∞–∑–ª–∏—á–∏—è:','{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
````
